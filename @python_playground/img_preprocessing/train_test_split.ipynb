```python
import os

os.chdir('data')
os.listdir()
```

```python
import glob

imgList = list(glob.glob('*.jpg'))
labelList = list(glob.glob('*.txt'))
data = list(zip(imgList, labelList))
data[-10:]
```

```python
len(imgList), len(labelList)
```

```python
import re

labelList = []
for name in imgList:
    labelList.append(re.sub("\..*", "", name) + ".txt")
list(zip(imgList, labelList))
```

```python
len(imgList), len(labelList)
```

```python
os.mkdir('./train')
os.mkdir('./test')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(imgList, labelList, test_size=0.3, random_state=42)
list(zip(X_train, y_train))
```

```python
len(X_train), len(X_test)
```

```python
pwd = os.getcwd()
pwd
```

```python
for name1, name2 in zip(X_train, y_train):
    try:
        os.rename(pwd+"\\"+name2, pwd+"\\train\\"+name2)
        os.rename(pwd+"\\"+name1, pwd+"\\train\\"+name1)
    except FileNotFoundError:
        continue

for name1, name2 in zip(X_test, y_test):
    try:
        os.rename(pwd+"\\"+name2, pwd+"\\test\\"+name2)
        os.rename(pwd+"\\"+name1, pwd+"\\test\\"+name1)
    except FileNotFoundError:
        continue
```

```python
f1 = open("train.txt","w")
f2 = open("test.txt", "w")

for name in X_train:
    f1.write("/data/train/"+name+'\n')
for name in X_test:
    f2.write("/data/test/"+name+'\n')

f1.close()
f2.close()
```
